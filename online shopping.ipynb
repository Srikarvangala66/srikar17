```python
#detecting outliers 
def  detect_outliers():
    count=1
    plt.figure(figsize=(18,28))
    for col in  dp.columns[:-1]:
        plt.subplot(6,6,count)
        plt.boxplot(dp[col])
        plt.title(col)
        count +=1
        
    plt.show()
```

```python
detect_outliers()
```

```python
#remove outliers 
def remove_outlier(dp_in, col_name):
    q1 = dp_in[col_name].quantile(0.25)
    q3 = dp_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    dp_out = dp_in.loc[(dp_in[col_name] > fence_low) & (dp_in[col_name] < fence_high)]
    return dp_out
```

```python
remove_outlier
```

```python
for col in dp.columns[:1]:
    dp= remove_outlier(dp,col)
```
